```python
#initialize empty lists
dates=[]
times=[]
agencies=[]
locs=[]
descrs=[]
subags=[]
latds=[]
longtds=[]
vehtypes=[]
vehyrs=[]
vehmakes=[]
vehmodels=[]
vehcolors=[]
violtypes=[]
races=[]
genders=[]
drvstates=[]
licstates=[]
arrestypes=[]
```

```python
for violatn in violatndata['data']:
    dates.append(violatn[9])
    times.append(violatn[10])
    agencies.append(violatn[11])
    subags.append(violatn[12])
    descrs.append(violatn[13])
    locs.append(violatn[14])
    latds.append(violatn[15])
    longtds.append(violatn[16])
    vehtypes.append(violatn[28])
    vehyrs.append(violatn[29])
    vehmakes.append(violatn[30])
    vehmodels.append(violatn[31])
    vehcolors.append(violatn[32])
    violtypes.append(violatn[33])
    races.append(violatn[37])
    genders.append(violatn[38])
    drvstates.append(violatn[40])
    licstates.append(violatn[41])
    arrestypes.append(violatn[42])
```

```python
print(len(dates))
print(len(times))
print(len(agencies))
print(len(licstates))
print(len(arrestypes))
```

```python
violatnlist = zip(dates,times,agencies,subags,descrs,locs,latds,longtds,vehtypes,vehyrs,vehmakes,
                 vehmodels,vehcolors,violtypes,races,genders,drvstates,licstates,arrestypes)
```

```python
colnames = ['Date','Time','Agency','Sub-agency','Description','Location','Latitude','Longitude','Vehicle Type',
           'Vehicle Year','Make','Model','Color','Type of Viol.','Race','Gender','State','License State','Arrest Type']
```
