```python
import pandas as pd

from sqlalchemy import create_engine, inspect
```

```python
#db_file = r'datasets/BoardGamesSQLitedatabase.db'
db_file = r'datasets/boardgames.sqlite'
engine = create_engine(f"sqlite:///{db_file}")
engine.table_names()
```

```python
engine.table_names()
```

```python
engine.has_table('BoardGames')
```

```python
inspector = inspect(engine)
```

```python
BGfields = inspector.get_columns('BoardGames')
```

```python
BG_df = pd.DataFrame.from_dict(BGfields)
BG_df
```

```python
Topics2fields = inspector.get_columns('bgg.topics')
Topics2df = pd.DataFrame.from_dict(Topics2fields)
Topics2df
```

```python
Docsfields = inspector.get_columns('bgg.ldaOut.top.documents')
```

```python
Docsdf = pd.DataFrame.from_dict(Docsfields)
Docsdf                                
```

```python
Termsfields = inspector.get_columns('bgg.ldaOut.top.terms')
Termsdf = pd.DataFrame.from_dict(Termsfields)
Termsdf
```
